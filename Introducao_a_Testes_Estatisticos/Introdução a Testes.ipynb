### Vamos ver o filme 1 

```python
filmes = pd.read_csv('dados/movies.csv')
filmes.query('movieId == 1')
```

```python
notas1 = notas.query('movieId == 1')
notas1.head()
```

```python
plt.figure(figsize=(12,6))
ax = sns.histplot(notas1.rating, kde = True, stat = 'density')
ax.set(xlabel = 'Notas', ylabel = 'Densidade', title = 'Distribuição de Notas do Filme Toy Story')

ax = ax 
```

```python
ax = sns.boxplot(data = notas1.rating, orient='h')
ax.set(xlabel = 'Nota')
ax.set_title('Distribuição das Notas do Filme Toy Story')
```

```python
notas1.rating.count()
```

```python
zconfint(notas1.rating)
```

```python
from statsmodels.stats.weightstats import ztest

ztest(notas1.rating, value = 3.4320503405352594)[1]   #p-value (segundo valor) bem menor que 0.05 
```

```python
temp[0:1]
```

```python
np.random.seed(75241)
temp = notas1.sample(frac = 1).rating

def calcula_teste(i):
    media = temp[0:i].mean()
    p = ztest(temp[0:i], value = 3.4320503405352594)[1]
    return(i, media, p)
    
valores = np.array([calcula_teste(i) for i in range(2, len(temp))])

plt.figure(figsize=(12,8))
plt.plot(valores[:,0] , valores[:,1], label = 'Média')
plt.plot(valores[:,0], valores[:,2], label = 'P-Value')
plt.hlines(y = 0.05, xmin = 2, xmax = len(temp), colors = 'r')
plt.legend()
```

## Comparação de dois conjuntos de amostras

```python
print(ztest(notas1.rating, notas.rating))
zconfint(notas1.rating, notas.rating)
```

```python
from scipy.stats import ttest_ind

ttest_ind(notas1.rating, notas.rating)
```

```python
descr_todas_as_notas = DescrStatsW(notas.rating)
descr_toy_story = DescrStatsW(notas1.rating)

comparacao = descr_todas_as_notas.get_compare(descr_toy_story)

comparacao.summary()#t-test
```

```python
comparacao.summary(use_t = False) #z-test
```
